# Extraction using PyMuPDF

```python


import fitz  # PyMuPDF

doc = fitz.open("../Document/HSC26-Bangla1st-Paper.pdf")
text = "\n".join([page.get_text() for page in doc])

print(text[:1000])
```

```python
text
```

```python
def clean_bangla_text(text):
    import re
    text = text.replace('\u200c', '').replace('\u200d', '')
    text = re.sub(r'[^\u0980-\u09FF\s.,!?।]', '', text)  # Bangla + punctuation only
    return text

normalized = clean_bangla_text(text)
```

```python
normalized
```

```python
import re

def sentence_split(text):
    return re.split(r'(?<=[।!?])\s+', text)

sentences = sentence_split(normalized)

def chunk_sentences(sentences, max_words=400):
    chunks, chunk, count = [], "", 0
    for sent in sentences:
        words = sent.split()
        if count + len(words) > max_words:
            chunks.append(chunk.strip())
            chunk, count = sent, len(words)
        else:
            chunk += " " + sent
            count += len(words)
    if chunk:
        chunks.append(chunk.strip())
    return chunks

chunks = chunk_sentences(sentences)
```

# Using Mistral OCR

```python
from dotenv import load_dotenv
load_dotenv('../.env', override=True)
```

```python
# Mistral OCR

import os
from mistralai import Mistral

api_key = os.environ["MISTRAL_API_KEY"]
client = Mistral(api_key=api_key)
```

```python
uploaded_pdf = client.files.upload(
    file={
        "file_name": "HSC26-Bangla1st-Paper.pdf",
        "content": open("../Document/HSC26-Bangla1st-Paper.pdf", "rb"),
    },
    purpose="ocr"
)  
```

```python
signed_url = client.files.get_signed_url(file_id=uploaded_pdf.id)
```

```python
ocr_response = client.ocr.process(
    model="mistral-ocr-latest",
    document={
        "type": "document_url",
        "document_url": signed_url.url,
    },
    include_image_base64=True
) 
```

```python
ocr_response
```

```python
ocr_response.pages[2]
```

```python
import base64

def data_uri_tobytes(data_uri):
    _, encoded = data_uri.split(',', 1)
    return base64.b64decode(encoded)

def export_image(image):
    parsed_image = data_uri_tobytes(image.image_base64)
    with open(image.id, 'wb') as file:
        file.write(parsed_image)

with open('parsed_text', 'w') as f:
    for page in ocr_response.pages:
        f.write(page.markdown)
        for image in page.images:
            export_image(image)
```

## Load Data using Langchain Loader

```python
# Load data
from langchain.document_loaders import UnstructuredMarkdownLoader
loader = UnstructuredMarkdownLoader("extracted_folder/parsed_text.markdown")
doc = loader.load()
len(doc)
```

## Indexing i.e. Splitting Data into chunks using RecursiveCharacterTextSplitter

```python
# Chunking the markdown doc
from langchain.text_splitter import RecursiveCharacterTextSplitter
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1300, chunk_overlap=300)
splits = text_splitter.split_documents(doc)
len(splits)
```

```python
splits[1]
```

```python
# Index
from langchain_google_genai import GoogleGenerativeAIEmbeddings
from langchain_community.vectorstores import Chroma

vectorstore = Chroma.from_documents(
    documents=splits,
    embedding=GoogleGenerativeAIEmbeddings(model="models/embedding-001")
)

retriever = vectorstore.as_retriever()
```

### Multi-Query

```python
# Prompt

from langchain.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain_google_genai import ChatGoogleGenerativeAI


# Multi Query: Different Perspectives
template = """You are an AI language model assistant. Your task is to generate exactly five (5) 
different versions of the given user question to retrieve relevant documents from a vector 
database. By generating multiple perspectives on the user question, your goal is to help
the user overcome some of the limitations of the distance-based similarity search. 
Provide these alternative questions separated by newlines. Original question: {question}"""
prompt_perspectives = ChatPromptTemplate.from_template(template)


generate_queries = (
    prompt_perspectives
    | ChatGoogleGenerativeAI(model="gemini-2.5-pro")
    | StrOutputParser()
    | (lambda x: x.split("\n"))
)
```

```python
# Filter documents retrieval
from langchain.load import dumps, loads

def get_unique_union(documents: list[list]):
    """ Unique union of retrieved docs """
    # Flatten list of lists, and convert each Document to string
    flattened_docs = [dumps(doc) for sublist in documents for doc in sublist]
    # Get unique documents
    unique_docs = list(set(flattened_docs))
    # Return
    return [loads(doc) for doc in unique_docs]

# Retrieve
question = "বিয়ের সময় কল্যাণীর প্রকৃত বয়স কত ছিল?"
retrieval_chain = generate_queries | retriever.map() | get_unique_union
docs = retrieval_chain.invoke({"question":question})
len(docs)
```

```python
docs[:4]
```

```python
# Generation

from operator import itemgetter

# Final layer of involving LLM
template = """Answer (only the answer, no extra other text) the following question based on this context, if not found answer that you didn't find the answer:

{context}

Question: {question}
"""

prompt = ChatPromptTemplate.from_template(template)

llm = ChatGoogleGenerativeAI(model="gemini-2.5-pro")

final_rag_chain = (
    {"context": retrieval_chain, 
     "question": itemgetter("question")} 
    | prompt
    | llm
    | StrOutputParser()
)

final_rag_chain.invoke({"question":question})
```
